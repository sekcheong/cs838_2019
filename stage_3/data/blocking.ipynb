```python
import py_entitymatching as em
import pandas as pd
```

```python
table_a = 'Walmart.csv'
table_b = 'IMDB.csv'
candidate_set = '777_apply_rules_ds.csv'

dfa = pd.read_csv(table_a)
dfb = pd.read_csv(table_b)
dfc = pd.read_csv(candidate_set)
```

```python
b = dfc.sample(50)
b = pd.merge(b, dfa, how='left', left_on='A_id', right_on='_id')
b = pd.merge(b, dfb, how='left', left_on='B_id', right_on='_id')
b
```

```python
dfc_merge = pd.merge(dfc, dfa, how='left', left_on='A_id', right_on='_id')
dfc_merge = pd.merge(dfc_merge, dfb, how='left', left_on='B_id', right_on='_id', suffixes=('_l', '_r'))
dfc_merge.head()
```

```python
thrs = 0.3
stop_words = set(['a', 'an', 'and', 'are', 'as', 'at', 'be', 'by', 'for', 'from', 'has', 'he',
                 'in', 'is', 'it', 'its', 'on', 'that', 'the', 'to', 'was', 'were', 'will', 'with'])

def name_block(row):
    l = row['name_l']
    r = row['name_r']
    set_l = set(l.strip().split(' '))-stop_words
    set_r = set(r.strip().split(' '))-stop_words
    if len(set_l & set_r) / len(set_l | set_r) >= thrs:
        return True
    else:
        return False
```

```python
len(dfc_merge)
reduced_dfc = dfc_merge[dfc_merge.apply(name_block, axis=1)]
len(reduced_dfc)
```

```python
b = reduced_dfc.sample(50)
b
```

```python
result_dfc = reduced_dfc[['A_id', 'B_id']]
result_dfc.head()
```

```python
result_dfc.to_csv('reduced_candidate.csv', index=0)
L = reduced_dfc.sample(400)
L.to_csv('L.csv', index=0)
```

```python
overlap_size = 1

def genre_block(row):
    l = row['genre_l']
    r = row['genre_r']
    if type(l)!=str or type(r)!=str:
        return True
    set_l = set(l.strip().split(' '))-stop_words if type(l)==str else set()
    set_r = set(r.strip().split(' '))-stop_words if type(r)==str else set()
    if len(set_l & set_r) >= overlap_size:
        return True
    else:
        return False
```

```python
len(reduced_dfc)
reduced_dfc2 = reduced_dfc[reduced_dfc.apply(genre_block, axis=1)]
len(reduced_dfc2)
```
